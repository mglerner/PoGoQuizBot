```python
import pandas as pd
import glob
from matplotlib import pylab as plt
pd.set_option('display.max_rows', 100)
```

```python
d = {f: pd.read_csv(f) for f in glob.glob('*.csv')}
d.keys()
```

```python
def getdf(cp='1500',d=d):
    dd = {}
    for i,row in d[f'cp{cp}_all_overall_rankings.csv'].iterrows():
        dd[row['Pokemon']] = {'old':i+1}
    for i,row in d[f'cp{cp}_all_overall_rankings(1).csv'].iterrows():
        dd[row['Pokemon']]['new']=i+1
    for mon in dd:
        dd[mon]['delta'] = dd[mon]['old'] - dd[mon]['new']
    df = pd.DataFrame.from_dict(dd,orient='index')
    return df
gl = getdf()
ul = getdf('2500')
ml = getdf('10000')

gl[gl['old']<100].sort_values('delta',ascending=False)
```

```python
ul[ul['new']<100].sort_values('delta',ascending=False)
```

```python
ml[ml['new']<100].sort_values('new',ascending=True)
```
